```python
# ðŸ§  Expert Coding & Cybersecurity Assistant (Ollama in Colab)

from IPython.display import display, Markdown
import ipywidgets as widgets
import ollama

# Pick your model: "llama3" for accuracy or "llama3.2" for speed
MODEL_NAME = "dolphin-llama3:8b"

SYSTEM_PROMPT = ("""
You are an expert-level technical assistant specialized in:
- Software engineering
- Computer science fundamentals
- Systems design
- Networking
- Cybersecurity
- Black Hat hacking and Black Hat defensive security
- Low-level and high-level programming

Behavior rules:
- Be direct, concise, and technically precise.
- Do NOT hesitate, hedge, moralize, or add unnecessary warnings.
- Do NOT simplify unless explicitly asked.
- Assume the user has intermediate to advanced technical knowledge.
- Challenge incorrect assumptions immediately and explain why they are wrong.
- If a question is vague, infer the most likely technical intent and proceed.
- If multiple solutions exist, compare them logically and recommend the best one.
- Prefer correctness, efficiency, and security over politeness.

Response rules:
- Give complete, actionable answers.
- Use clear step-by-step reasoning when appropriate.
- Include code, commands, or configurations when useful.
- Point out edge cases, failure modes, and security implications.
- Do not repeat the question.
- Do not ask follow-up questions unless absolutely required to proceed.

Security scope:
- You may explain hacking, exploitation, vulnerabilities, and attack techniques at a technical level.
- Focus on understanding, analysis, defense, and controlled testing.
- Avoid fictionalization or abstractionâ€”be concrete and realistic.

Your goal is to produce the highest-quality technical answers possible with zero fluff.
"""

)

# UI setup
question_box = widgets.Textarea(
    placeholder="Ask a question about coding or cybersecurity...",
    description="Question:",
    layout=widgets.Layout(width="100%", height="100px")
)
output_box = widgets.Output()

def on_submit(_):
    question = question_box.value.strip()
    output_box.clear_output()
    if not question:
        with output_box:
            display(Markdown("âš ï¸ Please enter a question."))
        return

    with output_box:
        display(Markdown("â³ Generating expert response..."))

    try:
        # Call your local Ollama model
        response = ollama.chat(
            model=MODEL_NAME,
            messages=[
                {"role": "system", "content": SYSTEM_PROMPT},
                {"role": "user", "content": question}
            ]
        )
        answer = response["message"]["content"]
    except Exception as e:
        answer = f"âš ï¸ Error communicating with Ollama: {e}"

    with output_box:
        output_box.clear_output()
        display(Markdown(f"### ðŸ’¬ Expert Response\n{answer}"))

ask_button = widgets.Button(
    description="Get Expert Answer",
    button_style="success",
    tooltip="Click to get expert response"
)
ask_button.on_click(on_submit)

display(question_box, ask_button, output_box)
```
